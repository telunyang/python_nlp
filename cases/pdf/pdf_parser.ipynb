```python
# 匯入模組
import pymupdf
```

```python
# 開啟 PDF 檔案
doc = pymupdf.open("./1706.03762v7.pdf")

# 讀取每一頁的文字
for page in doc:
    # 取得頁面文字
    text = page.get_text()

    # 印出頁碼與文字內容
    print(f"----- Page {page.number} -----")
    print(text.replace("\n", " "))
    print()
```
